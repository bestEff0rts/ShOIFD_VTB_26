````python
import pandas as pd

# Путь к файлу
file_path = "./data/opendata/data-20260323T0100-structure-20171205T0100.csv"

# Загрузка данных
try:
    df = pd.read_csv(file_path)
    print(f"Файл успешно загружен. Размер данных: {df.shape}")
except FileNotFoundError:
    print(f"Ошибка: файл не найден по пути {file_path}")
    raise
except Exception as e:
    print(f"Ошибка при загрузке файла: {e}")
    raise

# Базовые проверки
print(f"\nКоличество строк: {len(df)}")
print(f"Количество столбцов: {len(df.columns)}")
print(f"Типы данных столбцов:\n{df.dtypes}")
print(f"\nНаличие пропусков:\n{df.isnull().sum()}")

# Вывод первых 5 строк
print("\nПервые 5 строк данных:")
display(df.head(5))
````
