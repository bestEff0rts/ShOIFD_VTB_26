#### 4.2. Немного про библиотеку Pandas

[Pandas](https://pandas.pydata.org/) — это быстрый, мощный, гибкий и простой в использовании инструмент с открытым исходным кодом для анализа и обработки данных, построенный на языке программирования Python.

```python
# %% [markdown]
# # Анализ структуры данных с использованием Pandas
# 
# В этом ноутбуке демонстрируются 10 наиболее часто используемых функций библиотеки Pandas
# для работы с данными на примере файла со структурой данных.

# %% [markdown]
# ## Импорт необходимых библиотек и загрузка данных

# %%
import pandas as pd
import numpy as np
from IPython.display import display

# Загрузка данных из CSV файла
# Параметр sep=';' указывает разделитель, так как в данных используется точка с запятой
# encoding='utf-8' обеспечивает корректное чтение кириллических символов
df = pd.read_csv(
    './data/opendata/data-20260323T0100-structure-20171205T0100.csv',
    sep=',',
    encoding='utf-8'
)

# Выводим информацию о загруженных данных
print("Данные успешно загружены!")
print(f"Размер данных: {df.shape[0]} строк, {df.shape[1]} столбцов")

# %% [markdown]
# ## 1. head() - просмотр первых строк данных

# %%
def demonstrate_head():
    """
    Демонстрация функции head() - просмотр первых n строк DataFrame.
    Позволяет быстро оценить структуру данных, типы значений и наличие пропусков.
    """
    print("\n" + "="*50)
    print("1. Функция head() - просмотр первых 5 строк данных")
    print("="*50)
    
    # head() без параметров показывает первые 5 строк
    # Это самый быстрый способ ознакомиться с данными после загрузки
    display(df.head())
    
    # Можно указать произвольное количество строк
    print("\nПервые 3 строки данных:")
    display(df.head(3))

demonstrate_head()

# %% [markdown]
## 2. info() - информация о DataFrame

# %%
def demonstrate_info():
    """
    Демонстрация функции info() - получение сводной информации о DataFrame.
    Показывает типы данных, количество ненулевых значений и занимаемую память.
    """
    print("\n" + "="*50)
    print("2. Функция info() - информация о структуре данных")
    print("="*50)
    
    # info() выводит типы столбцов, количество ненулевых значений
    # и общую информацию о DataFrame
    # verbose=True показывает полную информацию по всем столбцам
    df.info(verbose=True, show_counts=True)

demonstrate_info()

# %% [markdown]
## 3. describe() - статистическое описание данных

# %%
def demonstrate_describe():
    """
    Демонстрация функции describe() - статистическое описание числовых данных.
    Возвращает основные статистические показатели: среднее, медиану, 
    стандартное отклонение, квартили, минимум и максимум.
    """
    print("\n" + "="*50)
    print("3. Функция describe() - статистическое описание данных")
    print("="*50)
    
    # describe() для числовых столбцов
    # Показывает основные статистические метрики
    display(df.describe())
    
    # Для категориальных данных можно использовать include='object'
    print("\nСтатистика для категориальных данных:")
    display(df.describe(include=['object']))

demonstrate_describe()

# %% [markdown]
## 4. groupby() - группировка и агрегация данных

# %%
def demonstrate_groupby():
    """
    Демонстрация функции groupby() - группировка данных для агрегации.
    Позволяет разбить данные на группы по одному или нескольким признакам
    и применить агрегирующие функции к каждой группе.
    """
    print("\n" + "="*50)
    print("4. Функция groupby() - группировка и агрегация данных")
    print("="*50)
    
    # Предполагаем, что в данных есть столбцы, подходящие для группировки
    # Выбираем первые 3 столбца для группировки (если они есть)
    group_columns = df.columns[:2].tolist() if len(df.columns) >= 2 else df.columns[:1]
    
    if len(group_columns) >= 1:
        # Группируем по первому столбцу и считаем количество записей в каждой группе
        print(f"Группировка по столбцу '{group_columns[0]}' (первые 10 групп):")
        grouped = df.groupby(group_columns[0]).size().reset_index(name='count')
        display(grouped.head(10))
        
        # Пример с несколькими агрегациями
        if len(df.columns) > 2 and pd.api.types.is_numeric_dtype(df[df.columns[2]]):
            print(f"\nАгрегация по '{group_columns[0]}': сумма, среднее и количество по '{df.columns[2]}'")
            agg_result = df.groupby(group_columns[0])[df.columns[2]].agg([
                ('сумма', 'sum'),
                ('среднее', 'mean'),
                ('количество', 'count')
            ]).reset_index()
            display(agg_result.head(10))
    else:
        print("Недостаточно столбцов для демонстрации группировки")

demonstrate_groupby()

# %% [markdown]
## 5. loc[] и iloc[] - индексация и выборка данных

# %%
def demonstrate_indexing():
    """
    Демонстрация функций loc[] и iloc[] - мощных инструментов для выборки данных.
    loc[] - выборка по меткам индексов и названиям столбцов
    iloc[] - выборка по целочисленным позициям
    """
    print("\n" + "="*50)
    print("5. Индексация с помощью loc[] и iloc[]")
    print("="*50)
    
    # iloc[] - выборка по целочисленным позициям
    print("iloc[] - выборка первых 3 строк и первых 2 столбцов:")
    display(df.iloc[:3, :2])
    
    # loc[] - выборка по меткам индексов и названиям столбцов
    print("\nloc[] - выборка строк 0-2 и столбцов по названиям:")
    if len(df.columns) >= 2:
        display(df.loc[0:2, df.columns[:2]])
    
    # Фильтрация с помощью логических условий
    print("\nФильтрация с помощью loc[]: строки где первый столбец не пустой (первые 5)")
    first_col = df.columns[0]
    mask = df[first_col].notna()
    display(df.loc[mask, :].head(5))

demonstrate_indexing()

# %% [markdown]
## 6. isnull() и dropna() - обработка пропущенных значений

# %%
def demonstrate_null_handling():
    """
    Демонстрация функций isnull(), notnull() и dropna() для работы с пропусками.
    Позволяют обнаружить и обработать отсутствующие значения в данных.
    """
    print("\n" + "="*50)
    print("6. Работа с пропущенными значениями: isnull(), dropna()")
    print("="*50)
    
    # Подсчет пропусков по каждому столбцу
    print("Количество пропущенных значений по столбцам:")
    missing_values = df.isnull().sum()
    display(missing_values[missing_values > 0].to_frame('пропуски'))
    
    # Процент пропусков
    missing_percent = (df.isnull().sum() / len(df)) * 100
    print("\nПроцент пропущенных значений по столбцам:")
    display(missing_percent[missing_percent > 0].to_frame('% пропусков'))
    
    # Удаление строк с пропусками (демонстрация)
    df_clean = df.dropna()
    print(f"\nПосле удаления всех строк с пропусками: {df_clean.shape[0]} строк (было {df.shape[0]})")
    
    # Удаление столбцов с большим количеством пропусков
    threshold = 50  # Удаляем столбцы с более чем 50% пропусков
    cols_to_drop = missing_percent[missing_percent > threshold].index
    if len(cols_to_drop) > 0:
        print(f"\nСтолбцы с >{threshold}% пропусков (будут удалены): {list(cols_to_drop)}")
        df_cols_clean = df.drop(columns=cols_to_drop)
        print(f"Размер после удаления столбцов: {df_cols_clean.shape}")

demonstrate_null_handling()

# %% [markdown]
## 7. value_counts() - подсчет уникальных значений

# %%
def demonstrate_value_counts():
    """
    Демонстрация функции value_counts() - подсчет частоты уникальных значений.
    Особенно полезна для категориальных данных и анализа распределений.
    """
    print("\n" + "="*50)
    print("7. Функция value_counts() - подсчет уникальных значений")
    print("="*50)
    
    # Поиск категориальных столбцов (тип object)
    categorical_cols = df.select_dtypes(include=['object']).columns
    
    if len(categorical_cols) > 0:
        # Демонстрация на первом категориальном столбце
        col_name = categorical_cols[0]
        print(f"Распределение значений в столбце '{col_name}':")
        value_counts_result = df[col_name].value_counts()
        display(value_counts_result.head(10))
        
        # С нормализацией (проценты)
        print(f"\nРаспределение в процентах (первые 10):")
        display(df[col_name].value_counts(normalize=True).head(10))
        
        # С учетом пропусков
        print(f"\nРаспределение с учетом пропусков:")
        display(df[col_name].value_counts(dropna=False).head(10))
    else:
        print("Категориальные столбцы не найдены")
        
        # Для числовых столбцов
        numeric_cols = df.select_dtypes(include=[np.number]).columns
        if len(numeric_cols) > 0:
            print(f"\nЧисловой столбец '{numeric_cols[0]}':")
            display(df[numeric_cols[0]].value_counts().head(10))

demonstrate_value_counts()

# %% [markdown]
## 8. merge() - объединение данных

# %%
def demonstrate_merge():
    """
    Демонстрация функции merge() - объединение двух DataFrame.
    Аналог SQL JOIN, позволяет комбинировать данные из разных источников.
    """
    print("\n" + "="*50)
    print("8. Функция merge() - объединение данных")
    print("="*50)
    
    # Создаем два небольших DataFrame для демонстрации
    # Если в основном df есть данные, создаем на их основе тестовые данные
    if len(df.columns) >= 2 and len(df) > 10:
        # Берем первые 10 строк и создаем две таблицы
        df1 = df.iloc[:10, :2].copy()
        df2 = df.iloc[5:15, :2].copy()
        
        print("Первый DataFrame (первые 5 строк):")
        display(df1.head())
        print("\nВторой DataFrame (первые 5 строк):")
        display(df2.head())
        
        # Внутреннее объединение (inner join)
        merged_inner = pd.merge(df1, df2, on=df1.columns[0], how='inner')
        print(f"\nВнутреннее объединение по столбцу '{df1.columns[0]}': {len(merged_inner)} записей")
        display(merged_inner.head())
        
        # Левое объединение (left join)
        merged_left = pd.merge(df1, df2, on=df1.columns[0], how='left')
        print(f"\nЛевое объединение: {len(merged_left)} записей")
        display(merged_left.head())
    else:
        # Демонстрация на созданных тестовых данных
        df_left = pd.DataFrame({
            'id': [1, 2, 3, 4],
            'name': ['A', 'B', 'C', 'D']
        })
        df_right = pd.DataFrame({
            'id': [2, 3, 4, 5],
            'value': [100, 200, 300, 400]
        })
        
        print("Левая таблица:")
        display(df_left)
        print("\nПравая таблица:")
        display(df_right)
        
        merged = pd.merge(df_left, df_right, on='id', how='inner')
        print("\nРезультат внутреннего объединения:")
        display(merged)

demonstrate_merge()

# %% [markdown]
## 9. apply() - применение функций к данным

# %%
def demonstrate_apply():
    """
    Демонстрация функции apply() - применение пользовательских функций.
    Позволяет гибко преобразовывать данные, применяя произвольные функции.
    """
    print("\n" + "="*50)
    print("9. Функция apply() - применение функций к данным")
    print("="*50)
    
    # Создаем копию небольшой части данных
    if len(df) > 0:
        sample_df = df.iloc[:10, :3].copy()
        
        # Применение функции к столбцу (если столбец числовой)
        numeric_cols = sample_df.select_dtypes(include=[np.number]).columns
        if len(numeric_cols) > 0:
            col_name = numeric_cols[0]
            print(f"Исходные данные столбца '{col_name}':")
            display(sample_df[[col_name]].head())
            
            # Применяем функцию для замены отрицательных значений на 0
            def replace_negative(x):
                """Заменяет отрицательные значения на 0"""
                if pd.isna(x):
                    return x
                return max(0, x)
            
            sample_df[f'{col_name}_clean'] = sample_df[col_name].apply(replace_negative)
            print("\nПосле применения функции (отрицательные заменены на 0):")
            display(sample_df[[col_name, f'{col_name}_clean']].head())
        
        # Применение функции ко всему DataFrame
        print("\nПрименение функции для подсчета количества символов в строковых столбцах:")
        string_cols = sample_df.select_dtypes(include=['object']).columns
        if len(string_cols) > 0:
            for col in string_cols[:1]:  # Берем первый строковый столбец
                sample_df[f'{col}_length'] = sample_df[col].apply(lambda x: len(str(x)) if pd.notna(x) else 0)
                print(f"\nДобавлен столбец с длиной строк для '{col}':")
                display(sample_df[[col, f'{col}_length']].head())
    else:
        print("Недостаточно данных для демонстрации")

demonstrate_apply()

# %% [markdown]
## 10. sort_values() - сортировка данных

# %%
def demonstrate_sorting():
    """
    Демонстрация функции sort_values() - сортировка данных.
    Позволяет упорядочить данные по одному или нескольким столбцам.
    """
    print("\n" + "="*50)
    print("10. Функция sort_values() - сортировка данных")
    print("="*50)
    
    # Сортировка по первому столбцу
    if len(df.columns) >= 1:
        first_col = df.columns[0]
        print(f"Сортировка по столбцу '{first_col}' (по возрастанию):")
        sorted_asc = df.sort_values(by=first_col, ascending=True)
        display(sorted_asc.head(5))
        
        print(f"\nСортировка по столбцу '{first_col}' (по убыванию):")
        sorted_desc = df.sort_values(by=first_col, ascending=False)
        display(sorted_desc.head(5))
        
        # Сортировка по нескольким столбцам
        if len(df.columns) >= 2:
            second_col = df.columns[1]
            print(f"\nСортировка по двум столбцам: '{first_col}' (возрастание), '{second_col}' (убывание):")
            sorted_multi = df.sort_values(by=[first_col, second_col], ascending=[True, False])
            display(sorted_multi.head(5))
        
        # Сортировка с игнорированием регистра
        if df[first_col].dtype == 'object':
            print(f"\nСортировка с игнорированием регистра (если применимо):")
            sorted_case = df.sort_values(by=first_col, key=lambda col: col.str.lower())
            display(sorted_case.head(5))

demonstrate_sorting()

# %% [markdown]
## Заключение

# %%
def summary():
    """
    Подводит итог использованным функциям Pandas.
    """
    print("\n" + "="*50)
    print("ИТОГ: 10 основных функций Pandas")
    print("="*50)
    
    functions_summary = {
        '1. head()': 'Просмотр первых строк данных',
        '2. info()': 'Информация о DataFrame (типы данных, пропуски)',
        '3. describe()': 'Статистическое описание данных',
        '4. groupby()': 'Группировка и агрегация данных',
        '5. loc/iloc': 'Индексация и выборка данных',
        '6. isnull/dropna': 'Обнаружение и обработка пропусков',
        '7. value_counts()': 'Подсчет частоты уникальных значений',
        '8. merge()': 'Объединение нескольких DataFrame',
        '9. apply()': 'Применение пользовательских функций',
        '10. sort_values()': 'Сортировка данных'
    }
    
    for func, description in functions_summary.items():
        print(f"{func:20} - {description}")
    
    print("\nЭти функции покрывают основные задачи EDA (Exploratory Data Analysis)")
    print("и являются фундаментом для более сложного анализа данных.")

summary()

# %% [markdown]
# ## Дополнительная информация
# 
# **Примечания по использованию:**
# - Все функции демонстрируются на реальных данных
# - Для работы в Jupyter Notebook используется `display()` для красивого вывода
# - Код написан с учетом возможных особенностей структуры данных
# - Обработка ошибок и проверка наличия данных встроены в каждую функцию
```

##### __4.3.1. Уникальные значения с Pandas__

```python
# -*- coding: utf-8 -*-
"""Получение уникальных значений поля SHORTNAME."""

import pandas as pd
from IPython.display import display

def get_unique_shortnames(df: pd.DataFrame) -> None:
    """
    Извлекает и отображает уникальные значения из поля SHORTNAME.
    
    Args:
        df: DataFrame с колонкой SHORTNAME
        
    Returns:
        None: Функция выводит результат напрямую
    """
    # Проверка наличия колонки
    if 'SHORTNAME' not in df.columns:
        print("Ошибка: Колонка 'SHORTNAME' отсутствует в данных")
        return
    
    # Получение уникальных значений
    unique_values = df['SHORTNAME'].unique()
    
    # Сортировка для удобства просмотра
    unique_values_sorted = sorted(unique_values)
    
    # Вывод результатов
    print(f"Всего уникальных значений: {len(unique_values_sorted)}")
    print("\nУникальные значения SHORTNAME:")
    print("-" * 50)
    
    # В интерактивном ноутбуке display удобнее для больших данных
    if len(unique_values_sorted) > 20:
        display(pd.DataFrame({
            'SHORTNAME': unique_values_sorted,
            '№': range(1, len(unique_values_sorted) + 1)
        }).set_index('№'))
    else:
        for i, value in enumerate(unique_values_sorted, 1):
            print(f"{i}. {value}")

# Демонстрация первых строк данных
print("Первые 5 строк данных:")
print("-" * 50)
display(df.head())

print("\n" + "=" * 50 + "\n")

# Основной вывод
get_unique_shortnames(df)

# Альтернативный однострочный вариант (закомментирован)
# unique_shortnames = df['SHORTNAME'].unique()
# display(pd.Series(sorted(unique_shortnames), name='Уникальные SHORTNAME'))
```

##### __4.3.2. Максимальный спред с Pandas__

```python
# -*- coding: utf-8 -*-
"""Расчет максимального спреда по всем ценным бумагам."""

import pandas as pd
from IPython.display import display

def convert_price_column(series: pd.Series) -> pd.Series:
    """
    Преобразует строковые значения цен в числовой формат.
    
    Args:
        series: Series с ценами в формате строки (с запятой как разделителем)
        
    Returns:
        Series с числовыми значениями
    """
    # Замена запятой на точку и преобразование в float
    return series.str.replace(',', '.').astype(float)

def calculate_max_spread(df: pd.DataFrame) -> None:
    """
    Рассчитывает максимальный спред (HIGH - LOW) по всем ценным бумагам.
    
    Args:
        df: DataFrame с колонками SHORTNAME, HIGH, LOW
        
    Returns:
        None: Функция выводит результат напрямую
    """
    # Проверка наличия необходимых колонок
    required_columns = ['SHORTNAME', 'HIGH', 'LOW']
    missing_columns = [col for col in required_columns if col not in df.columns]
    
    if missing_columns:
        print(f"Ошибка: Отсутствуют колонки: {missing_columns}")
        return
    
    # Создание копии для работы
    df_clean = df.copy()
    
    # Проверка и преобразование типов данных
    print("🔧 Проверка типов данных:")
    print(f"   HIGH тип: {df_clean['HIGH'].dtype}")
    print(f"   LOW тип: {df_clean['LOW'].dtype}")
    
    # Преобразование строк в числа, если необходимо
    if df_clean['HIGH'].dtype == 'object':
        print("   ⚠️  Преобразование HIGH из строки в число...")
        df_clean['HIGH'] = convert_price_column(df_clean['HIGH'])
    
    if df_clean['LOW'].dtype == 'object':
        print("   ⚠️  Преобразование LOW из строки в число...")
        df_clean['LOW'] = convert_price_column(df_clean['LOW'])
    
    # Расчет спреда для каждой строки
    df_clean['SPREAD'] = df_clean['HIGH'] - df_clean['LOW']
    
    # Проверка на отрицательные спреды (ошибки данных)
    negative_spreads = df_clean[df_clean['SPREAD'] < 0]
    if len(negative_spreads) > 0:
        print(f"\n⚠️  Найдено {len(negative_spreads)} записей с отрицательным спредом")
        print("   (LOW > HIGH - возможно, перепутаны колонки или ошибка данных)")
    
    # Поиск максимального спреда
    max_spread_idx = df_clean['SPREAD'].idxmax()
    max_spread_row = df_clean.loc[max_spread_idx]
    
    # Получение топ-10 максимальных спредов
    top_10_spreads = df_clean.nlargest(10, 'SPREAD')[
        ['SHORTNAME', 'HIGH', 'LOW', 'SPREAD']
    ]
    
    # Вывод результатов
    print("\n" + "=" * 70)
    print("АНАЛИЗ МАКСИМАЛЬНОГО СПРЕДА (HIGH - LOW)")
    print("=" * 70)
    
    print(f"\n📊 Максимальный спред по всем бумагам:")
    print(f"   Ценная бумага: {max_spread_row['SHORTNAME']}")
    print(f"   Максимальная цена (HIGH): {max_spread_row['HIGH']:.4f}")
    print(f"   Минимальная цена (LOW): {max_spread_row['LOW']:.4f}")
    print(f"   Спред: {max_spread_row['SPREAD']:.4f}")
    
    print(f"\n📈 Топ-10 бумаг с наибольшим спредом:")
    display(top_10_spreads)
    
    # Дополнительная статистика (только для положительных спредов)
    positive_spreads = df_clean[df_clean['SPREAD'] >= 0]['SPREAD']
    if len(positive_spreads) > 0:
        print(f"\n📐 Общая статистика по спредам (только положительные):")
        print(f"   Средний спред: {positive_spreads.mean():.4f}")
        print(f"   Медианный спред: {positive_spreads.median():.4f}")
        print(f"   Стандартное отклонение: {positive_spreads.std():.4f}")
    
    # Анализ по уникальным бумагам (агрегированный)
    print(f"\n🏢 Топ-5 бумаг по максимальному спреду (в разрезе бумаг):")
    max_spread_per_security = df_clean.groupby('SHORTNAME')['SPREAD'].max()
    top_5_securities = max_spread_per_security.nlargest(5)
    
    display(pd.DataFrame({
        'Максимальный спред': top_5_securities
    }))

# Демонстрация первых строк с HIGH и LOW
print("Первые 10 строк с ценами HIGH/LOW (исходный формат):")
display(df[['SHORTNAME', 'HIGH', 'LOW']].head(10))

print("\n" + "=" * 70 + "\n")

# Основной расчет
calculate_max_spread(df)

# Альтернативный упрощенный вариант для быстрой проверки
print("\n" + "=" * 70)
print("БЫСТРЫЙ ВАРИАНТ:")
print("=" * 70)

# Преобразование и расчет в одну строку
df['HIGH_NUM'] = df['HIGH'].str.replace(',', '.').astype(float)
df['LOW_NUM'] = df['LOW'].str.replace(',', '.').astype(float)
max_spread = (df['HIGH_NUM'] - df['LOW_NUM']).max()
max_idx = (df['HIGH_NUM'] - df['LOW_NUM']).idxmax()

print(f"Максимальный спред: {max_spread:.4f}")
print(f"Ценная бумага: {df.loc[max_idx, 'SHORTNAME']}")
print(f"HIGH: {df.loc[max_idx, 'HIGH']}, LOW: {df.loc[max_idx, 'LOW']}")
```

##### __4.3.3. Группировка данных с Pandas__

```python
import pandas as pd
import numpy as np

def clean_value_column(df):
    """
    Очищает и преобразует столбец VALUE в числовой формат.
    
    Args:
        df: DataFrame с исходными данными
    
    Returns:
        pd.DataFrame: DataFrame с очищенным столбцом VALUE
    """
    # Проверяем тип данных в столбце VALUE
    if df['VALUE'].dtype == 'object':
        print("\nОчистка столбца VALUE: преобразование строк в числа")
        
        # Заменяем запятые на точки и преобразуем в float
        df['VALUE'] = df['VALUE'].astype(str).str.replace(',', '.')
        df['VALUE'] = pd.to_numeric(df['VALUE'], errors='coerce')
        
        print(f"  Преобразовано строк в числа: {df['VALUE'].notna().sum()}")
        print(f"  Пропущено некорректных значений: {df['VALUE'].isna().sum()}")
    
    return df


def load_and_check_data(df):
    """
    Проверяет загруженные данные и выводит информацию о структуре.
    
    Args:
        df: DataFrame с исходными данными
    
    Returns:
        pd.DataFrame: Проверенный DataFrame
    """
    print("=" * 50)
    print("ПРОВЕРКА ДАННЫХ")
    print("=" * 50)
    
    # Вывод первых строк
    print("\nПервые 5 строк данных:")
    display(df.head())
    
    # Информация о размерности
    print(f"\nРазмер данных: {df.shape[0]} строк, {df.shape[1]} столбцов")
    
    # Проверка наличия необходимых столбцов
    required_columns = ['SECID', 'VALUE']
    missing_columns = [col for col in required_columns if col not in df.columns]
    
    if missing_columns:
        raise ValueError(f"Отсутствуют необходимые столбцы: {missing_columns}")
    
    # Очистка и преобразование столбца VALUE
    df = clean_value_column(df)
    
    # Проверка типов данных после очистки
    print(f"\nТип столбца VALUE после очистки: {df['VALUE'].dtype}")
    
    # Проверка на пропуски
    null_counts = df[required_columns].isnull().sum()
    if null_counts.any():
        print(f"\nПредупреждение: найдены пропуски в данных:")
        print(null_counts[null_counts > 0])
    
    # Удаляем строки с некорректными значениями VALUE
    initial_count = len(df)
    df = df.dropna(subset=['VALUE'])
    removed_count = initial_count - len(df)
    
    if removed_count > 0:
        print(f"\nУдалено строк с некорректным VALUE: {removed_count}")
    
    return df


def group_by_secid_and_sum_value(df):
    """
    Группирует данные по SECID и суммирует общий объем торгов VALUE.
    
    Args:
        df: DataFrame с данными
    
    Returns:
        pd.DataFrame: Результат группировки с общим объемом торгов по каждому SECID
    """
    print("\n" + "=" * 50)
    print("ГРУППИРОВКА ДАННЫХ")
    print("=" * 50)
    
    # Проверяем, что VALUE числовой
    if not np.issubdtype(df['VALUE'].dtype, np.number):
        raise TypeError(f"Столбец VALUE не числовой: {df['VALUE'].dtype}")
    
    # Группировка по SECID и суммирование VALUE
    result = df.groupby('SECID', as_index=False)['VALUE'].sum()
    
    # Переименование столбца для наглядности
    result = result.rename(columns={'VALUE': 'TOTAL_VALUE'})
    
    # Сортировка по убыванию общего объема (безопасная сортировка)
    result = result.sort_values('TOTAL_VALUE', ascending=False, na_position='last')
    
    # Проверяем тип результата
    print(f"Тип TOTAL_VALUE после группировки: {result['TOTAL_VALUE'].dtype}")
    
    return result


def display_results(result_df):
    """
    Выводит результаты группировки в удобном формате.
    
    Args:
        result_df: DataFrame с результатами группировки
    """
    print("\n" + "=" * 50)
    print("РЕЗУЛЬТАТЫ ГРУППИРОВКИ")
    print("=" * 50)
    
    print(f"\nНайдено уникальных SECID: {len(result_df)}")
    
    # Проверяем, что TOTAL_VALUE числовой перед выводом статистики
    if np.issubdtype(result_df['TOTAL_VALUE'].dtype, np.number):
        print("\nТоп-10 SECID по общему объему торгов:")
        display(result_df.head(10))
        
        # Базовая статистика
        print("\nСтатистика по общему объему торгов:")
        stats = result_df['TOTAL_VALUE'].describe()
        print(f"  Среднее: {stats['mean']:,.2f}")
        print(f"  Медиана: {stats['50%']:,.2f}")
        print(f"  Максимум: {stats['max']:,.2f}")
        print(f"  Минимум: {stats['min']:,.2f}")
        print(f"  Общий объем всех торгов: {result_df['TOTAL_VALUE'].sum():,.2f}")
    else:
        print("Ошибка: TOTAL_VALUE не числовой тип данных")
        print(f"Текущий тип: {result_df['TOTAL_VALUE'].dtype}")
        print("\nПервые 5 строк для диагностики:")
        display(result_df.head())


# Основной код выполнения
def main_analysis(df):
    """
    Основная функция анализа: группировка по SECID и подсчет общего объема VALUE.
    
    Args:
        df: Исходный DataFrame
    
    Returns:
        pd.DataFrame: Результат группировки
    """
    try:
        # Проверка данных
        df_checked = load_and_check_data(df)
        
        # Группировка и суммирование
        result = group_by_secid_and_sum_value(df_checked)
        
        # Вывод результатов
        display_results(result)
        
        return result
        
    except Exception as e:
        print(f"\nОшибка при выполнении анализа: {e}")
        print("\nДополнительная информация для диагностики:")
        print(f"Столбцы в данных: {df.columns.tolist()}")
        print(f"Типы данных:\n{df.dtypes}")
        raise


# Выполнение анализа
# Если данные уже загружены в переменную df
if 'df' in locals():
    result_df = main_analysis(df)
else:
    print("Ошибка: переменная 'df' не найдена")
    print("Убедитесь, что данные загружены в переменную df")
```

##### __4.3.3. Визуализация с Pandas__

```python
# Базовые импорты
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Настройка стиля графиков
plt.style.use('default')

def load_and_explore_data(df):
    """
    Выполняет первичный анализ данных и построение гистограммы DURATION.
    
    Args:
        df: DataFrame с данными, содержащий колонку DURATION
    """
    # Отображение первых строк данных
    print("Первые 5 строк данных:")
    display(df.head())
    
    print("\n" + "="*50)
    print("ИНФОРМАЦИЯ О ДАННЫХ")
    print("="*50)
    
    # Базовая информация о данных
    print(f"\nРазмер данных: {df.shape[0]} строк x {df.shape[1]} столбцов")
    
    # Проверка наличия колонки DURATION
    if 'DURATION' not in df.columns:
        print("ОШИБКА: Колонка 'DURATION' не найдена в данных!")
        print(f"Доступные колонки: {df.columns.tolist()}")
        return
    
    # Статистика по DURATION
    print(f"\nСтатистика по показателю DURATION:")
    print(f"  Минимальное значение: {df['DURATION'].min():.4f}")
    print(f"  Максимальное значение: {df['DURATION'].max():.4f}")
    print(f"  Среднее значение: {df['DURATION'].mean():.4f}")
    print(f"  Медиана: {df['DURATION'].median():.4f}")
    print(f"  Стандартное отклонение: {df['DURATION'].std():.4f}")
    
    # Проверка на пропущенные значения
    missing_count = df['DURATION'].isna().sum()
    if missing_count > 0:
        print(f"\nПРЕДУПРЕЖДЕНИЕ: Пропущенных значений DURATION: {missing_count} ({missing_count/len(df)*100:.2f}%)")
    
    # Построение гистограммы
    print("\n" + "="*50)
    print("ПОСТРОЕНИЕ ГИСТОГРАММЫ")
    print("="*50)
    
    # Создание фигуры с двумя подграфиками
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # Гистограмма 1: базовая
    ax1.hist(df['DURATION'].dropna(), bins=30, edgecolor='black', alpha=0.7, color='steelblue')
    ax1.set_xlabel('DURATION', fontsize=12)
    ax1.set_ylabel('Частота', fontsize=12)
    ax1.set_title('Гистограмма распределения DURATION\n(30 интервалов)', fontsize=14)
    ax1.grid(True, alpha=0.3)
    
    # Гистограмма 2: с ядерной оценкой плотности
    ax2.hist(df['DURATION'].dropna(), bins=30, density=True, edgecolor='black', 
             alpha=0.7, color='seagreen', label='Гистограмма')
    ax2.set_xlabel('DURATION', fontsize=12)
    ax2.set_ylabel('Плотность', fontsize=12)
    ax2.set_title('Нормированная гистограмма DURATION', fontsize=14)
    ax2.grid(True, alpha=0.3)
    
    # Добавление линии среднего значения
    mean_val = df['DURATION'].mean()
    ax2.axvline(mean_val, color='red', linestyle='--', linewidth=2, 
                label=f'Среднее: {mean_val:.4f}')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Дополнительный анализ: выбросы
    print("\n" + "="*50)
    print("АНАЛИЗ ВЫБРОСОВ")
    print("="*50)
    
    # Метод межквартильного размаха (IQR)
    Q1 = df['DURATION'].quantile(0.25)
    Q3 = df['DURATION'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df['DURATION'] < lower_bound) | (df['DURATION'] > upper_bound)]
    
    print(f"Нижняя граница (Q1 - 1.5*IQR): {lower_bound:.4f}")
    print(f"Верхняя граница (Q3 + 1.5*IQR): {upper_bound:.4f}")
    print(f"Количество выбросов: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)")
    
    if len(outliers) > 0:
        print(f"\nПримеры выбросов (первые 5):")
        display(outliers[['DURATION']].head())
    
    return fig

# Основной код выполнения
print("="*60)
print("АНАЛИЗ ДАННЫХ: ПОСТРОЕНИЕ ГИСТОГРАММЫ DURATION")
print("="*60)

# Вызов функции анализа
fig = load_and_explore_data(df)

print("\nАнализ завершен успешно!")
```
